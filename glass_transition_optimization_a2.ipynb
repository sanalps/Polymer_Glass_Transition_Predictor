# Additional Descriptor model

```python
df = pd.read_csv ('/content/gdrive/MyDrive/git_projects/glass_transition/glass_transition/glass_transitiona1_add_desc.csv')
df
```

```python
X = df.drop([' Polymer Name','Smile', 'Tg_k' ], axis='columns')
X
```

```python
y = df['Tg_k']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)
```

```python
from lazypredict.Supervised import LazyRegressor
```

```python
#modulus lazy regression
reg_mod = LazyRegressor(verbose=0, ignore_warnings=True, custom_metric = None)
### fitting data in LazyClassifier
models, predictions = reg_mod.fit(X_train, X_test, y_train, y_test)
### lets check which model did better on rubber Dataset
print(models)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
etree_reg = ExtraTreesRegressor()
etree_reg.fit(X_train, y_train)
```

```python
y_predict = etree_reg.predict(X_test)
```

```python
#import numpy as np
from sklearn.metrics import mean_squared_error
forest_mse = mean_squared_error(y_predict, y_test)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
etree_reg.get_params()
```

# Random search optimization

```python
from sklearn.model_selection import RandomizedSearchCV
#Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 2000, num = 10)]
#Number of features to consider at evey split
max_features = ['auto', 'sqrt']
#maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
#minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
#minimm number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
#method of selecting samples for training each tree
bootstrap = [True, False]

#create the random grid
random_grid = {'n_estimators' : n_estimators, 'max_features' : max_features, 
              'max_depth' : max_depth, 'min_samples_split' : min_samples_split,
              'min_samples_leaf' : min_samples_leaf, 'bootstrap' : bootstrap }
print(random_grid)
```

```python
etree_reg2 = ExtraTreesRegressor()
# use 3 fold cross validation
# search 100 different combinations
etree_random = RandomizedSearchCV(estimator=etree_reg2, param_distributions=random_grid, n_iter=100, cv = 3,
                             verbose=2, random_state=42)
etree_random.fit(X_train, y_train)
```

```python
etree_random.best_params_
```

```python
etree_reg.get_params()
```

```python
etree_rand_best = etree_random.best_estimator_
```

```python
y_predict = etree_rand_best.predict(X_test)
```

```python
#import numpy as np
from sklearn.metrics import mean_squared_error
forest_mse = mean_squared_error(y_predict, y_test)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```
